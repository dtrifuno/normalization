```python
import whisperx

device = "cpu"
audio_file = "../Speech/Data/Obicni Luge/Season 4/s04e01_planinari.wav"
batch_size = 16 # reduce if low on GPU mem
compute_type = "int8" # change to "int8" if low on GPU mem (may reduce accuracy)
```

```python
model = whisperx.load_model("large-v2", device, compute_type=compute_type)
```

```python
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio, batch_size=batch_size)
print(result["segments"]) # before alignment
```

```python
audio_file1 = "../Speech/Data/Obicni Luge/Season 4/s04e02_malahovi.wav"
audio1 = whisperx.load_audio(audio_file1)
result1 = model.transcribe(audio1, batch_size=batch_size)
#print(result1["segments"]) # before alignment
```

```python
import pandas as pd

df = pd.read_csv("segments.csv", dtype=str, na_filter=False)
```

```python
#result['segments']
for i, line in enumerate(df["0"]):
    result1['segments'][i]['text'] = line
```

```python
result1['segments']
```

```python
model_a, metadata = whisperx.load_align_model('mk', device, model_name="dtrifuno/wav2vec2-xls-r-300m-mk")
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)
```

```python
result1 = whisperx.align(result1["segments"], model_a, metadata, audio1, device, return_char_alignments=False)
```

```python
from whisperx.utils import WriteSRT

with open("b.srt", "w", encoding="utf-8") as srt:
    writesrt = WriteSRT(".")
    writesrt.write_result(result=result1, file=srt, options={"max_line_width": None, "max_line_count": None, "highlight_words": False})
```
