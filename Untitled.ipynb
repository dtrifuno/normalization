```python
%autoreload 2
from normalization.fleurs import process

def process_batch(batch):
    batch["transcription"] = process(batch["transcription"])
    return batch

data = data.map(process_batch)
```

```python
VALID_CHARS = 'абвгдежзиклмнопрстуфхцчшѐѓѕјљњќџ '
valid_pattern = r'[абвгдежзиклмнопрстуфхцчшѐѓѕјљњќџ ]*'

bad_lines = {}
for i, line in enumerate(data['transcription']):
    if re.fullmatch(valid_pattern, line) is not None:
        continue
    if line in bad_lines:
        bad_lines[line].append(i)
    else:
        bad_lines[line] = [i]
        
print(len(bad_lines))
for line, idxs in bad_lines.items():
    print(idxs)
    print(line)
    print()
```

```python
VALID_CHARS = 'абвгдежзиклмнопрстуфхцчшѐѓѕјљњќџ '
valid_pattern = r'[^абвгдежзиклмнопрстуфхцчшѐѓѕјљњќџ ].*'

s = set()
for k in data['transcription']:
    m = re.search(valid_pattern, k)
    if m is not None:
        s |= set(k)

sorted(list(s))
```
